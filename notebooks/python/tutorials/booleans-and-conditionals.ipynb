# ブール値（Booleans）

Pythonには `bool` という型の変数があります。取りうる値は `True` と `False` の2つです。

````python
x = True
print(x)
print(type(x))
````

コード中に直接 `True` や `False` を書くよりも、通常は**ブール演算子**からブール値を取得します。これらは「はい/いいえ」の質問に答える演算子です。以下でいくつかの演算子を見ていきましょう。

## 比較演算

| 演算子        | 説明                              || 演算子        | 説明                                 |
|---------------|-----------------------------------||---------------|--------------------------------------|
| ``a == b``    | ``a`` は ``b`` と等しい           || ``a != b``    | ``a`` は ``b`` と等しくない          |
| ``a < b``     | ``a`` は ``b`` より小さい         || ``a > b``     | ``a`` は ``b`` より大きい            |
| ``a <= b``    | ``a`` は ``b`` 以下               || ``a >= b``    | ``a`` は ``b`` 以上                  |



````python
def can_run_for_president(age):
    """指定された年齢の人はアメリカ大統領選に出馬できるか？"""
    # アメリカ合衆国憲法では35歳以上である必要がある
    return age >= 35

print("Can a 19-year-old run for president?", can_run_for_president(19))
print("Can a 45-year-old run for president?", can_run_for_president(45))
````

比較は多くの場合、期待通りに動作します。

````python
3.0 == 3
````

しかし、時には注意が必要な場合もあります。

````python
'3' == 3
````

比較演算子はすでに学んだ算術演算子と組み合わせることで、事実上無限の数学的テストを表現できます。例えば、2で割った余りが1かどうかを確認することで、数が奇数かどうかを判定できます：

````python
def is_odd(n):
    return (n % 2) == 1

print("Is 100 odd?", is_odd(100))
print("Is -1 odd?", is_odd(-1))
````

比較を行うときは `=` ではなく `==` を使うことを忘れないでください。`n == 2` と書くと n の値について質問しています。`n = 2` と書くと n の値を変更しています。

## ブール値の組み合わせ
ブール値は「かつ（and）」、「または（or）」、「でない（not）」という標準的な概念を使って組み合わせることができます。Pythonでは、そのまま ``and``、``or``、``not`` というキーワードを使います。

これらを使って、`can_run_for_president` 関数をより正確にしてみましょう。

````python
def can_run_for_president(age, is_natural_born_citizen):
    """指定された年齢と市民権の状態で、アメリカ大統領選に出馬できるか？"""
    # アメリカ合衆国憲法では、出生による市民権を持ち、*かつ*35歳以上である必要がある
    return is_natural_born_citizen and (age >= 35)

print(can_run_for_president(19, True))
print(can_run_for_president(55, False))
print(can_run_for_president(55, True))
````

さて、この式の値を推測できますか？

````python
True or True and False
````

（「output」ボタンをクリックして答えを確認してください）

これに答えるには、演算の優先順位を理解する必要があります。

例えば、`and` は `or` よりも先に評価されます。そのため、上の最初の式は `True` になります。左から右に評価した場合、まず `True or True`（結果は `True`）を計算し、次にその結果と `False` の `and` を取って、最終的な値が `False` になるはずでした。

[演算子の優先順位を暗記](https://docs.python.org/3/reference/expressions.html#operator-precedence)しようとすることもできますが、括弧を積極的に使う方が安全です。これはバグの防止に役立つだけでなく、コードを読む人に対してあなたの意図を明確にします。

例えば、次の式を考えてみましょう：

```python
prepared_for_weather = have_umbrella or rain_level < 5 and have_hood or not rain_level > 0 and is_workday
```

私が言いたいのは、今日の天気に対して安全だということです...
- 傘を持っている場合...
- または雨がそれほど強くなく、フードを持っている場合...
- それ以外でも、雨が降っていて*かつ*仕事の日でなければ大丈夫

しかし、このPythonコードは読みにくいだけでなく、バグもあります。括弧を追加することで両方の問題に対処できます：

```python
prepared_for_weather = have_umbrella or (rain_level < 5 and have_hood) or not (rain_level > 0 and is_workday)
```

読みやすさを向上させるために、さらに括弧を追加することもできます：

```python
prepared_for_weather = have_umbrella or ((rain_level < 5) and have_hood) or (not (rain_level > 0 and is_workday))
```

上で説明した3つの部分の構造を強調するために、複数行に分けることもできます：

```python
prepared_for_weather = (
    have_umbrella 
    or ((rain_level < 5) and have_hood) 
    or (not (rain_level > 0 and is_workday))
)
```

# 条件分岐（Conditionals）

ブール値は*条件文*と組み合わせたときに最も有用になります。条件文では ``if``、``elif``、``else`` というキーワードを使います。

条件文は*if-then文*とも呼ばれ、ブール条件の値に基づいてどのコードを実行するかを制御できます。以下に例を示します：

````python
def inspect(x):
    if x == 0:
        print(x, "is zero")
    elif x > 0:
        print(x, "is positive")
    elif x < 0:
        print(x, "is negative")
    else:
        print(x, "is unlike anything I've ever seen...")

inspect(0)
inspect(-15)
````

``if`` と ``else`` キーワードは他の言語でもよく使われます。Pythonに特徴的なキーワードは ``elif`` で、「else if」の短縮形です。
条件節では ``elif`` と ``else`` ブロックはオプションです。また、``elif`` 文は好きなだけ含めることができます。

特に、コロン（`:`）と空白を使って別々のコードブロックを示していることに注目してください。これは関数を定義するときと同様です。関数のヘッダーは `:` で終わり、次の行は4つのスペースでインデントされます。後続のインデントされた行はすべて関数の本体に属し、インデントされていない行に出会うと関数の定義が終わります。

````python
def f(x):
    if x > 0:
        print("Only printed when x is positive; x =", x)
        print("Also only printed when x is positive; x =", x)
    print("Always printed, regardless of x's value; x =", x)

f(1)
f(0)
````

## ブール値への変換

`int()` はものを整数に変換し、`float()` はものを浮動小数点数に変換します。同様に、Pythonにはものをブール値に変換する `bool()` 関数があると聞いても驚かないでしょう。

````python
print(bool(1)) # 0以外のすべての数値はTrueとして扱われる
print(bool(0))
print(bool("asf")) # 空文字列""以外のすべての文字列はTrueとして扱われる
print(bool(""))
# 一般的に、空のシーケンス（文字列、リスト、まだ学んでいないタプルなど）は
# 「偽（falsey）」で、それ以外は「真（truthy）」
````

ブール値でないオブジェクトを `if` 条件やブール値が期待される場所で使うことができます。Pythonは暗黙的にそれらを対応するブール値として扱います：

````python
if 0:
    print(0)
elif "spam":
    print("spam")
````
