# はじめに

Pythonで変数を作成するたびに、その変数には対応するデータ型を持つ値が格納されます。整数、浮動小数点数、ブール値、文字列など、さまざまなデータ型があり、このレッスンではそれらすべてを扱います。（これは利用可能なデータ型のほんの一部にすぎません。辞書、セット、リスト、タプルなど、さらに多くのデータ型があります。）

データ型は重要です。なぜなら、データ型によってどのような操作ができるかが決まるからです。例えば、2つの浮動小数点数を割り算することはできますが、2つの文字列を割り算することはできません。つまり、`12.0/2.0` は意味がありますが、`"cat"/"dog"` は意味がありません。

エラーを避けるためには、操作が扱っているデータ型に適合していることを確認する必要があります。

# 整数（int）

整数は小数部分を持たない数値で、正の数（`1`、`2`、`3`、...）、負の数（`-1`、`-2`、`-3`、...）、またはゼロ（`0`）のいずれかです。

以下のコードセルでは、変数 `x` に整数を設定しています。次に `type()` を使用してデータ型を確認します。括弧の中に変数名を渡すだけです。

```python
x = 14
print(x)
print(type(x))
```

上の出力で、`<class 'int'>` は**整数（int）**データ型を指しています。

# 浮動小数点数（float）

浮動小数点数は小数部分を持つ数値です。小数点以下に多くの桁を持つことができます。

```python
nearly_pi = 3.141592653589793238462643383279502884197169399375105820974944
print(nearly_pi)
print(type(nearly_pi))
```

分数を使って浮動小数点数を指定することもできます。

```python
almost_pi = 22/7
print(almost_pi)
print(type(almost_pi))
```

分数に特に便利な関数の1つが `round()` 関数です。指定した小数点以下の桁数に数値を丸めることができます。

```python
# 小数点以下5桁に丸める
rounded_pi = round(almost_pi, 5)
print(rounded_pi)
print(type(rounded_pi))
```

小数点付きの数値を書くたびに、Pythonはそれを浮動小数点数データ型として認識します。

例えば、`1.`（または `1.0`、`1.00` など）は浮動小数点数として認識されます。これらの数値には技術的には小数部分がないにもかかわらず、そうなります！

```python
y_float = 1.
print(y_float)
print(type(y_float))
```

# ブール値（bool）

ブール値は `True` または `False` の2つの値のいずれかを表します。以下のコードセルでは、`z_one` に値 `True` を持つブール値を設定しています。

```python
z_one = True
print(z_one)
print(type(z_one))
```

次に、`z_two` に値 `False` を持つブール値を設定します。

```python
z_two = False
print(z_two)
print(type(z_two))
```

ブール値は式の真偽値を表すために使われます。`1 < 2` は真の文なので、`z_three` は `True` の値を取ります。

```python
z_three = (1 < 2)
print(z_three)
print(type(z_three))
```

同様に、`5 < 3` は偽の文なので、`z_four` は `False` の値を取ります。

```python
z_four = (5 < 3)
print(z_four)
print(type(z_four))
```

`not` を使用してブール値の値を反転させることができます。つまり、`not True` は `False` と等しく、`not False` は `True` になります。

```python
z_five = not z_four
print(z_five)
print(type(z_five))
```

ブール値は次のレッスンで条件と条件文について学ぶ際に重要になります。

# 文字列（str）

文字列データ型は、引用符で囲まれた文字（アルファベット、句読点、数字、記号など）の集まりです。文字列は一般的にテキストを表すために使用されます。

```python
w = "Hello, Python!"
print(w)
print(type(w))
```

`len()` を使って文字列の長さを取得できます。`"Hello, Python!"` の長さは14です。スペース、カンマ、感嘆符を含む14文字があるためです。長さの計算時に引用符は含まれないことに注意してください。

```python
print(len(w))
```

特殊な文字列の1つに空文字列があり、その長さはゼロです。

```python
shortest_string = ""
print(type(shortest_string))
print(len(shortest_string))
```

数値を引用符で囲むと、文字列データ型になります。

```python
my_number = "1.12321"
print(my_number)
print(type(my_number))
```

浮動小数点数に変換可能な文字列がある場合、`float()` を使用できます。

これは常に機能するわけではありません！例えば、`"10.43430"` や `"3"` は浮動小数点数に変換できますが、`"Hello, Python!"` は浮動小数点数に変換できません。

```python
also_my_number = float(my_number)
print(also_my_number)
print(type(also_my_number))
```

2つの数値（浮動小数点数または整数）を足し算できるのと同様に、2つの文字列を足し算することもできます。その結果、2つの元の文字列を連結した、より長い文字列が得られます。

```python
new_string = "abc" + "def"
print(new_string)
print(type(new_string))
```

2つの文字列で引き算や割り算はできないことに注意してください。また、2つの文字列同士の掛け算もできませんが、文字列に整数を掛けることはできます。この場合も結果は文字列で、元の文字列を指定された回数だけ連結したものになります。

```python
newest_string = "abc" * 3
print(newest_string)
print(type(newest_string))
```

文字列に浮動小数点数を掛けることはできないことに注意してください！試みるとエラーが返されます。

```python
will_not_work = "abc" * 3.
```

エラーメッセージの中で、「sequence」は文字列 `"abc"` を指し、「non-int of type 'float'」は浮動小数点数（`3.`）を指しています。つまり、このエラーメッセージは「文字列に浮動小数点数を掛けることはできません」と言い換えることができます。
