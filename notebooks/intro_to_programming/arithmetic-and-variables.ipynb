# 出力（Printing）

コンピュータに実行させることができる最もシンプルで（かつ最も重要な！）タスクの1つは、メッセージを出力することです。

Pythonでは、`print()`と書き、括弧の中に引用符（クォーテーションマーク）で囲んだメッセージを入れることで、コンピュータにメッセージの出力を指示します。以下では、コンピュータに `Hello, world!` というメッセージを出力するように指示しています。

```python
print("Hello, world!")
```

コードはボックス（**コードセル**と呼ばれます）の中にあり、コンピュータの応答（コードの**出力**と呼ばれます）はボックスの下に表示されます。ご覧の通り、コンピュータは私たちが望んだメッセージを出力しました。

# 算術演算（Arithmetic）

# コメント

コードが何をしているかを注釈するために**コメント**を使用します。これは他の人がコードを理解するのに役立ちますし、しばらく自分のコードを見ていなかった場合にも役立ちます。これまでに書いたコードは非常に短いものでしたが、たくさんのコードを書いた場合には注釈がより重要になります。

例えば、次のコードセルでは3に2を掛けます。また、コードの上に何をしているかを説明するコメント（`# Multiply 3 by 2`）を追加しています。

```python
# 3に2を掛ける
print(3 * 2)
```

ある行がコメントでありPythonコードではないことをPythonに知らせるには、最初の文字としてハッシュ記号（`#`）を書く必要があります。

Pythonがハッシュ記号を発見し、その行がコメントであると認識すると、コンピュータはその行を完全に無視します。これは重要です。なぜなら、英語や日本語（またはその他の言語）と同じように、Pythonには従わなければならない非常に厳密なルールがあるからです。しかし、Pythonは人間の聞き手よりも厳密であり、コードを適切に解釈できなければエラーを出します。

この例を下のコードセルで見ることができます。コメント内のテキストは有効なPythonコードではないため、ハッシュ記号を削除するとPythonは適切に解釈できずにエラーを出します。

```python
3に2を掛ける
```

# 変数（Variables）

これまでは、コードを使って計算を行い、その結果を出力してきましたが、結果はどこにも保存されていません。しかし、結果を保存して後で使いたいと思う場面が想像できるでしょう。そのためには**変数**を使う必要があります。

## 変数の作成

次のコードセルでは `test_var` という名前の変数を作成し、そこに4と5を足して得られる値を割り当てます。

その後、変数に割り当てられた値である9を出力します。

```python
# test_varという変数を作成し、4+5の値を割り当てる
test_var = 4 + 5

# test_varの値を出力する
print(test_var)
```

一般的に、変数を扱うには、まず使いたい名前を決める必要があります。変数名は短く説明的であるのが理想です。また、いくつかの要件を満たす必要があります：
- 空白を含めることはできません（例：`test var`は不可）
- アルファベット、数字、アンダースコア（_）のみを含めることができます（例：`test_var!`は不可）
- アルファベットまたはアンダースコアで始まる必要があります（例：`1_var`は不可）

次に、変数を作成するには、`=`（イコール記号）を使って持たせたい値を割り当てる必要があります。

変数に割り当てられた値は、`print()` を使用し括弧内に変数名を入れることでいつでも確認できます。

使い続けるうちに、Python変数の適切な名前の選び方が自然とわかってくるでしょう。今は違和感があっても全く問題ありませんし、一番の学習方法はたくさんのPythonコードを見ることです！

## 変数の操作

以前の値を上書きすることで、変数に割り当てられた値をいつでも変更することができます。

下のコードセルでは、`my_var` の値を3から100に変更しています。

```python
# 新しい変数に3の値を設定する
my_var = 3

# my_varに割り当てられた値を出力する
print(my_var)

# 変数の値を100に変更する
my_var = 100

# my_varに割り当てられた新しい値を出力する
print(my_var)
```

一般的に、コードセル内で変数を定義すると、それ以降のすべてのコードセルでもその変数にアクセスできることに注意してください。例えば、次のコードセルを使って、`my_var`（上のコードセルで定義）と `test_var`（このチュートリアルの前半で定義）の値にアクセスします。

```python
print(my_var)
print(test_var)
```

次のコードセルは、Pythonに `my_var` の現在の値を3増やすように指示します。

これを行うには、以前と同じように `my_var =` を使う必要があります。そして以前と同様に、変数に割り当てたい新しい値は `=` の右側に書きます。

```python
# 値を3増やす
my_var = my_var + 3

# my_varに割り当てられた値を出力する
print(my_var)
```

## 複数の変数の使用

コードで複数の変数を使用するのは一般的なことです。これは、複数の入力を使って長い計算を行うときに特に便利です。

次のコードセルでは、4年間の秒数を計算します。この計算では5つの変数（入力値）を使用します。

```python
# 変数を作成する
num_years = 4
days_per_year = 365 
hours_per_day = 24
mins_per_hour = 60
secs_per_min = 60

# 4年間の秒数を計算する
total_secs = secs_per_min * mins_per_hour * hours_per_day * days_per_year * num_years
print(total_secs)
```

上で計算したように、4年間には126,144,000秒あります。

変数を使用せずに、単に `60 * 60 * 24 * 365 * 4` としてこの計算を行うことも*可能*ですが、読みやすくないため、ミスがないかを確認するのがはるかに難しくなります。変数（`num_years`、`days_per_year` など）を使用すると、計算の各部分を適切に把握しやすくなり、間違いを確認して修正するのが簡単になります。

入力値が変更される可能性がある場合には、変数の使用が特に便利です。例えば、うるう年を考慮して、1年の日数を365から365.25に変更して見積もりをわずかに改善したいとします。その場合、他の変数を一切変更することなく、`days_per_year` に割り当てられた値を変更して再計算を行うことができます。

```python
# うるう年を含めるように更新する
days_per_year = 365.25

# 4年間の秒数を計算する
total_secs = secs_per_min * mins_per_hour * hours_per_day * days_per_year * num_years
print(total_secs)
```

**注**: 数字の最後に `.0` が追加されていることにお気づきかもしれません。一見不要に思えるかもしれませんが、これは2回目の計算で小数部分を持つ数字（365.25）を使用したことによります。対して1回目の計算では小数部分がない数字のみを掛け合わせました。これについては、データ型を扱うレッスン3で詳しく学びます。

## デバッグ（Debugging）

変数を扱う際によくあるエラーの1つが、誤ってタイポ（打ち間違い）をしてしまうことです。例えば、`hours_per_day` を `hours_per_dy` と書いてしまうと、Pythonは `NameError: name 'hours_per_dy' is not defined`（名前 'hours_per_dy' が定義されていません）というメッセージのエラーを出します。

```python
print(hours_per_dy)
```

このような `NameError` が出た場合、それは「定義されていない」とされている変数のスペルを確認すべきだというサインです。スペルを修正するだけでエラーを解決できます。

```python
print(hours_per_day)
```
