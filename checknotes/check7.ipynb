## 問題4: 関数を引数に渡す

次の `apply_twice` 関数は、関数 `fn` と値 `x` を受け取り、`fn` を `x` に2回適用した結果を返します。

`apply_twice` はそのまま使い、引数として渡す関数 `add_five`（引数に5を足す関数）を定義してください。

**期待される結果**:
- `apply_twice(add_five, 10)` → `20`（10 + 5 + 5）

```python
def apply_twice(fn, x):
    return fn(fn(x))

# add_five を定義してください


print(apply_twice(add_five, 10))
```
