# 入門

このレッスンでは、関数（Functions）を使ってコードを整理する方法を学びます。**関数**は、特定のタスクを実行するために設計されたコードのブロックです。後でわかるように、関数を使うと同じような計算を複数回おこなう際にコードを複製せずに済みます。

# 関数の基礎：簡単な例

まずは簡単な関数の例から始めましょう。以下の `add_three()` 関数は、任意の数字を受け取り、それに3を足して結果を返します。

```python
# Define the function
def add_three(input_var):
    output_var = input_var + 3
    return output_var
```

すべての関数は、**ヘッダー（header）**と**ボディ（body）**の2つの部分で構成されています。

![](https://storage.googleapis.com/kaggle-media/learn/images/gu0AWhK.png)

## ヘッダー
関数の**ヘッダー**は、関数の名前とその引数（argument）を定義します。
- すべての関数ヘッダーは `def` から始まります。これは、これから関数を定義することをPythonに伝えます。
- 例では、関数の名前は `add_three` です。
- 例では、引数は `input_var` です。**引数**とは、関数への入力として使用される変数の名前です。引数は常に関数名のすぐ後ろにある括弧で囲まれます。（*注：関数には引数がない場合もあれば、複数の引数がある場合もあります。このレッスンの後半でいくつか例を見ていきます。*）
- すべての関数において、関数の引数を囲む括弧の後にはコロン `:` を付ける必要があります。

## ボディ
関数の**ボディ**は、関数が実行する作業を指定します。
- 関数のボディにあるすべてのコード行は、正確に4つのスペースでインデント（字下げ）されている必要があります。スペースキーを4回押すか、キーボードの「Tab」ボタンを1回押すことで行うことができます。（*Pythonについてさらに学ぶにつれて、4つ以上のスペースでインデントする必要が出てくるかもしれませんが、それについてはこのコースの後半で詳しく学びます。*）
- 関数は、インデントされたすべての行を上から下へ実行することでその作業を行います。
    - 例にある `input_var` のように、引数を入力として受け取ります。
    - 関数は、計算 `output_var = input_var + 3` を用いて、新しい変数 `output_var` を作成します。
    - そして最後のコード行である**return（戻り値）文**が、関数の出力として `output_var` の値を単に返します。

上のコードセルは関数を定義しているだけであり、実行はしていません。関数のボディの詳細は、実際に関数を実行する次のコードセルの後にさらに意味がわかるようになるでしょう。

# 関数の実行（呼び出し）方法

関数を実行することを、関数を「呼び出す（calling）」とも表現します。

下のコードセルでは、入力値として `10` を指定して関数を実行します。そして、関数の出力に設定される新しい変数 `new_number` を定義しています。

```python
# Run the function with 10 as input
new_number = add_three(10)

# Check that the value is 13, as expected
print(new_number)
```

![](https://storage.googleapis.com/kaggle-media/learn/images/hlUbxQE.png)

詳細については以下の通りです：
- `add_three(10)` は、`input_var` の値として `10` を提供し、`add_three()` 関数を呼び出したときに出力として得られる値です。関数が実行されると、そのボディ内のすべてのコードが上から下へと実行されます：
  - まず `output_var = input_var + 3` を計算し、`output_var = 13` に設定されます。
  - 最後のコード行がreturn文で、`output_var` の値である `13` を返します。
- `new_number = add_three(10)` と設定することで、`new_number = 13` と設定したことになります。

**注**: このチュートリアルで何気なく `add_three()` 関数に言及するとき、関数名の後に空の閉じ括弧を使用しています。これは、人々が一般的にPythonコードの説明を書く方法と一致しており、空の括弧は、変数や他のPythonオブジェクトではなく関数を参照していることを明確にするためのものです。関数に引数がある場合でも、これらの括弧は常に空であるべきです。

# 関数の命名

上の例では、関数の名前はあらかじめ選ばれていました。自分で関数に名前を付けるときは、小文字のみを使用し、スペースの代わりにアンダースコアで単語を区切るようにしてください。

関数の命名は時間が経てば自然に感じられるようになりますし、最初は違和感があるのが普通です。Pythonのコードをたくさん見ることが一番の学習方法です。

# より複雑な例

基本がわかったところで、より長い計算の例に進みましょう。

友人が税引き後の週給を計算するのを手伝っているとします。
- 友人は12%の税率区分にいます（言い換えれば、給与の12%が税金として引かれ、88%だけを持ち帰ることになります）
- 友人は時給制で、時給15ドルで支払われます。

以下の関数は、働いた時間数に基づいて給与を計算します。前の例よりもコード行数やコメントが多いため、少し複雑になっています。上の例と同様に、関数には単一の引数（`num_hours`）があります。関数のボディでは：
- `num_hours` の値を使用して、新しい変数 `pay_pretax` の値を指定します。
- `pay_pretax` の値を使用して、新しい変数 `pay_aftertax` の値を指定します。
- `pay_aftertax` 変数の値を返します。

```python
def get_pay(num_hours):
    # Pre-tax pay, based on receiving $15/hour
    pay_pretax = num_hours * 15
    # After-tax pay, based on being in 12% tax bracket
    pay_aftertax = pay_pretax * (1 - .12)
    return pay_aftertax
```

この関数は、最初の関数を呼び出したのと同じ方法で呼び出します。次のコードセルは、40時間働いた場合に基づいて給与を計算します。（税引き後で528ドルです。）

```python
# Calculate pay based on working 40 hours
pay_fulltime = get_pay(40)
print(pay_fulltime)
```

異なる労働時間に基づいて給与を素早く計算するには、関数に異なる数字を提供するだけです。例えば、友人が32時間働いたとします。（すると、422.40ドルになります。）

```python
pay_parttime = get_pay(32)
print(pay_parttime)
```

関数を書いたおかげで、計算のコードをすべて書き直すことなく、異なる時間の給与を計算することができます。

関数はコードのエラーを避けるのに役立ち、また多くの時間を節約してくれます。一般的に、コーディングの際はできるだけ書く量を少なくすることを目指すべきです。なぜなら、計算を毎回打ち込むたびに、タイプミスやエラーを誤って持ち込んでしまう可能性があるからです。

# 変数の「スコープ（適用範囲）」

関数のボディ内で定義された変数は、関数の外からはアクセスできません。例えば、次のコードセルは `pay_aftertax` が関数の中にしか存在しないため、エラーになります。

```python
print(pay_aftertax)
```

`pay_pretax` や `num_hours` を出力しようとした場合も同じエラーになります。このため、関数から何らかの情報が必要な場合は、それが関数の最後にある return文 に含まれていることを確認する必要があります。

変数がアクセスできるコードの領域のことを、変数の**スコープ**と呼びます。関数のボディのように _関数の内部_ で定義された変数（たとえば `pay_aftertax`）は、その関数だけの「ローカルスコープ」を持ちます。しかし、これまでに見てきたように、_すべての関数の外_ で定義された変数（たとえば `pay_parttime`）は「グローバルスコープ」を持ち、どこからでもアクセスすることができます。

# 複数の引数を持つ関数

ここまで、1つの引数だけを持つ関数を定義する方法を学びました。複数の引数を持つ関数を定義するには、関数のヘッダーにある括弧の中により多くの引数を追加し、それらをカンマで区切るだけです。

以下の `get_pay_with_more_inputs()` 関数を使ってこれを行います。この関数は、次の3つの引数に基づいて週給を計算します：
- `num_hours` - 1週間に働いた時間数
- `hourly_wage` - 時給（ドル/時間）
- `tax_bracket` - 税金として引かれる給与の割合

```python
def get_pay_with_more_inputs(num_hours, hourly_wage, tax_bracket):
    # Pre-tax pay
    pay_pretax = num_hours * hourly_wage
    # After-tax pay
    pay_aftertax = pay_pretax * (1 - tax_bracket)
    return pay_aftertax
```

そして関数を呼び出すためには、各入力につき１つの値を提供し、またカンマで区切る必要があります。

以下のコードセルでは、40時間働き、時給24ドルで、22%の税率区分の人の税引き後給与を計算します。

```python
higher_pay_aftertax = get_pay_with_more_inputs(40, 24, .22)
print(higher_pay_aftertax)
```

次のコードセルは、`get_pay(40)` を実行したときと同じ結果になります。なぜなら `hourly_wage` が15に設定され、`tax_bracket` が12%に設定されており、これは `get_pay` を設計した方法と一致しているからです。

```python
same_pay_fulltime = get_pay_with_more_inputs(40, 15, .12)
print(same_pay_fulltime)
```

この新しい関数 `get_pay_with_more_inputs()` をどう使用するかにもよりますが、オリジナルの `get_pay()` 関数よりも多くのケースに対応できるため、より有用といえます。新しい関数では、ユーザーが時給と税率を誤って仮定してしまう可能性がある代わりに、正しい値を指定できるようになっています。しかし、もし時給と税率が変わる必要がないことがわかっているのであれば、新しい関数は必要以上に複雑なだけです。一般的に関数を定義する際は、使用ケースに基づいてどれだけの柔軟性が必要かを考慮する必要があります。

# 引数を持たない関数

引数を持たず、return文を持たない関数を定義することも可能であることに注意してください。下のコードセルの `print_hello()` 関数がその一例です。

```python
# Define the function with no arguments and with no return
def print_hello():
    print("Hello, you!")
    print("Good morning!")
    
# Call the function
print_hello()
```

このコースは短く、プログラミング入門として意図されているため、このチュートリアルで関数について知るべきことすべてを網羅するわけではありません。関数については、[Pythonコース](http://www.kaggle.com/learn/python)でさらに詳しく学ぶことになります。
