# はじめに

データサイエンスを行う際には、データを効率的に扱えるように整理する方法が必要です。Pythonには、リスト、セット、辞書、タプルなど、データを保持するための多くのデータ構造があります。このチュートリアルでは、Pythonのリストの使い方を学びます。

# モチベーション

[Petal to the Metalコンペティション](https://www.kaggle.com/c/tpu-getting-started)では、花の画像だけからその種類を分類することが目標です。（これはコンピュータビジョンにおける一般的なタスクで、画像分類と呼ばれます。）この目標に向けて、データに含まれる花の種名を整理するとしましょう。

一つの方法は、名前をPythonの文字列として整理することです。

```python
flowers = "pink primrose,hard-leaved pocket orchid,canterbury bells,sweet pea,english marigold,tiger lily,moon orchid,bird of paradise,monkshood,globe thistle"

print(type(flowers))
print(flowers)
```

さらに良い方法は、同じデータをPythonのリストで表現することです。リストを作成するには、角括弧（`[`、`]`）を使い、各要素をカンマで区切ります。リスト内のすべての要素はPythonの文字列なので、それぞれ引用符で囲まれています。

```python
flowers_list = ["pink primrose", "hard-leaved pocket orchid", "canterbury bells", "sweet pea", "english marigold", "tiger lily", "moon orchid", "bird of paradise", "monkshood", "globe thistle"]

print(type(flowers_list))
print(flowers_list)
```

一見すると、情報をPythonの文字列で表現してもリストで表現しても、あまり違いがないように見えます。しかしこれから見ていくように、リストを使うとより簡単にできるタスクがたくさんあります。例えば、リストを使うと以下のことが簡単になります：
- 指定した位置（1番目、2番目、3番目など）の要素を取得する
- 要素の数を確認する
- 要素を追加・削除する

# リスト

## 長さ

`len()`を使うと、任意のリストの要素数を数えることができます。`len`は「length（長さ）」の略です。括弧の中にリストの名前を指定するだけです。

```python
# リストには10個の要素がある
print(len(flowers_list))
```

## インデックス

リスト内の位置（1番目、2番目、3番目など）に基づいて、任意の要素を参照することができます。これを**インデックス**と呼びます。

Pythonはゼロベースのインデックスを使用することに注意してください。つまり：
- リストの最初の要素を取得するには、0を使います
- リストの2番目の要素を取得するには、1を使います
- リストの最後の要素を取得するには、リストの長さより1小さい数を使います

```python
print("最初の要素:", flowers_list[0])
print("2番目の要素:", flowers_list[1])

# リストの長さは10なので、最後の要素は9で参照する
print("最後の要素:", flowers_list[9])
```

**補足**: 上のコードセルでは、1つの`print()`で複数の項目（`"最初の要素:"`のようなPython文字列と、`flowers_list[0]`のようなリストの値）を表示していることに気づいたかもしれません。Pythonで1つのコマンドで複数のものを表示するには、カンマで区切るだけで済みます。

## スライス

リストの一部分（例えば、最初の3つの要素や最後の2つの要素）を取り出すこともできます。これを**スライス**と呼びます。例えば：
- 最初の`x`個の要素を取得するには、`[:x]`を使います
- 最後の`y`個の要素を取得するには、`[-y:]`を使います

```python
print("最初の3つの要素:", flowers_list[:3])
print("最後の2つの要素:", flowers_list[-2:])
```

上で見たように、リストをスライスすると、新しい短いリストが返されます。


## 要素の削除

`.remove()`を使ってリストから要素を削除します。削除したい要素を括弧の中に指定します。

```python
flowers_list.remove("globe thistle")
print(flowers_list)
```

## 要素の追加

`.append()`を使ってリストに要素を追加します。追加したい要素を括弧の中に指定します。

```python
flowers_list.append("snapdragon")
print(flowers_list)
```

## リストは文字列だけではない

ここまでは、リスト内の各要素が文字列であるリストのみを扱ってきました。しかし、リストにはブール値、整数、浮動小数点数など、あらゆるデータ型の要素を含めることができます。

例として、2000年4月の最初の1週間における、ある小売店でのハードカバー書籍の売上を考えてみましょう。

```python
hardcover_sales = [139, 128, 172, 139, 191, 168, 170]
```

ここで、`hardcover_sales`は整数のリストです。文字列の場合と同様に、長さの取得、個々の要素の取り出し、リストの拡張などを行うことができます。

```python
print("リストの長さ:", len(hardcover_sales))
print("インデックス2の要素:", hardcover_sales[2])
```

`min()`で最小値、`max()`で最大値を取得することもできます。

```python
print("最小値:", min(hardcover_sales))
print("最大値:", max(hardcover_sales))
```

リスト内のすべての要素を合計するには、`sum()`を使います。

```python
print("1週間の総売上冊数:", sum(hardcover_sales))
```

リストのスライスを使って同様の計算をすることもできます。次のコードセルでは、最初の5日間の合計（`sum(hardcover_sales[:5])`）を求め、それを5で割って最初の5日間の1日あたりの平均売上冊数を計算しています。

```python
print("最初の5日間の平均売上冊数:", sum(hardcover_sales[:5])/5)
```
