# 文字列（string）

Pythonが特に優れている分野の一つが文字列の操作です。
このセクションでは、Pythonの組み込み文字列メソッドとフォーマット機能を紹介します。

このような文字列操作のパターンは、データサイエンスの仕事でもよく登場します。

## 文字列の構文

これまでのレッスンでたくさんの文字列の例を見てきましたが、おさらいしておきましょう。Pythonの文字列は、シングルクォート（`'`）でもダブルクォート（`"`）でも定義できます。どちらも機能的には同じです。

```python
x = 'Pluto is a planet'
y = "Pluto is a planet"
x == y
```

ダブルクォートは、文字列にシングルクォート文字（例えばアポストロフィ）が含まれる場合に便利です。

同様に、ダブルクォートを含む文字列を作りたい場合は、シングルクォートで囲むと簡単です：

```python
print("Pluto's a planet!")
print('My dog is named "Pluto"')
```

シングルクォートで囲んだ文字列の中にシングルクォート文字を入れようとすると、Pythonは混乱してしまいます：

```python
'Pluto's a planet!'
```

バックスラッシュを使ってシングルクォートを「エスケープ」することで修正できます。

```python
'Pluto\'s a planet!'
```

以下の表は、バックスラッシュ文字の重要な使い方をまとめたものです。

| 入力する内容 | 得られる結果 | 例                    | `print(例)`                  |
|--------------|----------------|--------------------------------------------------------|
| `\'`         | `'`            | `'What\'s up?'`         | `What's up?`                 |  
| `\"`         | `"`            | `"That's \"cool\""`     | `That's "cool"`              |  
| `\\`         | `\`            |  `"Look, a mountain: /\\"` |  `Look, a mountain: /\`  |
| `\n`        |   <br/>      |   `"1\n2 3"`                       |   `1`<br/>`2 3`              |

最後のシーケンス `\n` は*改行文字*を表します。Pythonに新しい行を開始させます。

```python
hello = "hello\nworld"
print(hello)
```

さらに、Pythonのトリプルクォート構文を使うと、文字列に改行を文字通り含めることができます（特殊な '\n' シーケンスを使う代わりに、キーボードで 'Enter' を押すだけです）。関数のドキュメントとして使うdocstringですでに見ましたが、文字列を定義したい場所ならどこでも使えます。

```python
triplequoted_hello = """hello
world"""
print(triplequoted_hello)
triplequoted_hello == hello
```

`print()` 関数は、キーワード引数 `end` にデフォルト値の `'\n'` 以外の値を指定しない限り、自動的に改行文字を追加します：

```python
print("hello")
print("world")
print("hello", end='')
print("pluto", end='')
```

## 文字列はシーケンス

文字列は文字のシーケンス（並び）と考えることができます。リストに対してできることのほとんどは、文字列に対してもできます。

```python
# インデックス
planet = 'Pluto'
planet[0]
```

```python
# スライス
planet[-3:]
```

```python
# この文字列の長さは？
len(planet)
```

```python
# ループで回すこともできます
[char+'! ' for char in planet]
```

しかし、リストとの大きな違いは、文字列は*イミュータブル（不変）*であるということです。変更することができません。

```python
planet[0] = 'B'
# planet.appendも動作しません
```

## 文字列のメソッド

`list` と同様に、`str` 型にも非常に便利なメソッドがたくさんあります。ここではいくつかの例を紹介します。

```python
# すべて大文字に
claim = "Pluto is a planet!"
claim.upper()
```

```python
# すべて小文字に
claim.lower()
```

```python
# 部分文字列の最初のインデックスを検索
claim.index('plan')
```

```python
claim.startswith(planet)
```

```python
# 感嘆符がないのでFalse
claim.endswith('planet')
```

### 文字列とリストの変換：`.split()` と `.join()`

`str.split()` は文字列を小さな文字列のリストに変換します。デフォルトでは空白で分割します。1つの大きな文字列から単語のリストを得るのに非常に便利です。

```python
words = claim.split()
words
```

空白以外で分割したいこともあるでしょう：

```python
datestr = '1956-01-31'
year, month, day = datestr.split('-')
```

`str.join()` は逆の方向に変換します。文字列のリストを、呼び出した文字列を区切り文字として使って1つの長い文字列に結合します。

```python
'/'.join([month, day, year])
```

```python
# Unicode文字を文字列リテラルに直接入れることもできます :)
' 👏 '.join([word.upper() for word in words])
```

### `.format()` で文字列を構築する

Pythonでは `+` 演算子を使って文字列を連結できます。

```python
planet + ', we miss you.'
```

文字列でないオブジェクトを含めたい場合は、事前に `str()` を呼び出す必要があります。

```python
position = 9
planet + ", you'll always be the " + position + "th planet to me."
```

```python
planet + ", you'll always be the " + str(position) + "th planet to me."
```

これでは読みにくく、入力も面倒です。`str.format()` の出番です。

```python
"{}, you'll always be the {}th planet to me.".format(planet, position)
```

ずっとすっきりしましたね！「フォーマット文字列」に対して `.format()` を呼び出し、挿入したいPythonの値は `{}` プレースホルダーで表現します。

`position` を int から変換するために `str()` を呼ぶ必要すらなかったことに注目してください。`format()` がそれを処理してくれます。

`format()` がこれだけのことしかできなくても十分便利ですが、実はもっと*たくさんの*ことができます。その一端を紹介します：

```python
pluto_mass = 1.303 * 10**22
earth_mass = 5.9722 * 10**24
population = 52910390
#         小数点以下2桁      小数点以下3桁、パーセント形式     カンマ区切り
"{} weighs about {:.2} kilograms ({:.3%} of Earth's mass). It is home to {:,} Plutonians.".format(
    planet, pluto_mass, pluto_mass / earth_mass, population,
)
```

```python
# format()の引数を0から始まるインデックスで参照する
s = """Pluto's a {0}.
No, it's a {1}.
{0}!
{1}!""".format('planet', 'dwarf planet')
print(s)
```

`str.format` だけで短い本が書けるほどですので、ここで止めておきます。さらに詳しくは [pyformat.info](https://pyformat.info/) と [公式ドキュメント](https://docs.python.org/3/library/string.html#formatstrings) をご覧ください。

# 辞書（Dictionaries）

辞書は、キーと値を対応付けるためのPythonの組み込みデータ構造です。

```python
numbers = {'one':1, 'two':2, 'three':3}
```

この場合、`'one'`、`'two'`、`'three'` が**キー**で、1、2、3 がそれぞれ対応する値です。

値にはリストや文字列のインデックスと同様の角括弧構文でアクセスします。

```python
numbers['one']
```

同じ構文を使って、新しいキーと値のペアを追加できます。

```python
numbers['eleven'] = 11
numbers
```

既存のキーに関連付けられた値を変更することもできます。

```python
numbers['one'] = 'Pluto'
numbers
```

Pythonには、前のチュートリアルで見たリスト内包表記と似た構文の*辞書内包表記*があります。

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
planet_to_initial = {planet: planet[0] for planet in planets}
planet_to_initial
```

`in` 演算子は、辞書のキーに含まれているかどうかを教えてくれます。

```python
'Saturn' in planet_to_initial
```

```python
'Betelgeuse' in planet_to_initial
```

辞書に対するforループは、キーに対して繰り返します。

```python
for k in numbers:
    print("{} = {}".format(k, numbers[k]))
```

`dict.keys()` と `dict.values()` を使って、それぞれすべてのキーまたはすべての値のコレクションにアクセスできます。

```python
# すべての頭文字を取得し、アルファベット順にソートし、スペース区切りの文字列にする
' '.join(sorted(planet_to_initial.values()))
```

非常に便利な `dict.items()` メソッドを使うと、辞書のキーと値を同時に反復処理できます。（Pythonの用語では、**アイテム**はキーと値のペアを指します。）

```python
for planet, initial in planet_to_initial.items():
    print("{} begins with \"{}\"".format(planet.rjust(10), initial))
```

辞書のメソッドの完全な一覧を読むには、以下の「output」ボタンをクリックして完全なヘルプページを表示するか、[公式オンラインドキュメント](https://docs.python.org/3/library/stdtypes.html#dict)を確認してください。

```python
help(dict)
```
