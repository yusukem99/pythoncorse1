# ループ（繰り返し処理）

ループは、同じコードを繰り返し実行するための仕組みです。以下に例を示します：

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
for planet in planets:
    print(planet, end=' ') # すべて同じ行に表示
```

``for`` ループでは以下を指定します：
- 使用する変数名（この場合は `planet`）
- ループで回す値の集合（この場合は `planets`）

これらを ``in`` というキーワードでつなぎます。

``in`` の右側のオブジェクトは、反復（イテレーション）をサポートする任意のオブジェクトです。基本的に、ものの集まりと考えられるものであれば、おそらくループで回すことができます。リストに加えて、タプルの要素に対しても反復できます：

```python
multiplicands = (2, 2, 2, 3, 3, 5)
product = 1
for mult in multiplicands:
    product = product * mult
product
```

文字列の各文字に対してもループを回すことができます：

```python
s = 'steganograpHy is the practicE of conceaLing a file, message, image, or video within another fiLe, message, image, Or video.'
msg = ''
# sの中の大文字をすべて1文字ずつ出力する
for char in s:
    if char.isupper():
        print(char, end='')        
```

### range()

`range()` は数値のシーケンスを返す関数です。ループを書くときにとても便利です。

例えば、ある処理を5回繰り返したい場合：

```python
for i in range(5):
    print("Doing important work. i =", i)
```

## ``while`` ループ
Pythonのもう1つのループは ``while`` ループで、ある条件が満たされるまで繰り返します：

```python
i = 0
while i < 10:
    print(i, end=' ')
    i += 1 # iの値を1増やす
```

``while`` ループの引数はブール文として評価され、その文が False になるまでループが実行されます。

# リスト内包表記（List Comprehensions）

リスト内包表記はPythonで最も愛され、最も独特な機能の1つです。理解するには、いくつかの例を見るのが一番でしょう：

```python
squares = [n**2 for n in range(10)]
squares
```

リスト内包表記を使わずに同じことをする場合は次のようになります：

```python
squares = []
for n in range(10):
    squares.append(n**2)
squares
```

`if` 条件を追加することもできます：

```python
short_planets = [planet for planet in planets if len(planet) < 6]
short_planets
```

（SQLに馴染みがある方は、これをSQLの「WHERE」句のようなものだと考えることができるかもしれません。）

以下は、`if` 条件でフィルタリングしつつ、*さらに*ループ変数に何らかの変換を適用する例です：

```python
# str.upper()は文字列の全大文字バージョンを返す
loud_short_planets = [planet.upper() + '!' for planet in planets if len(planet) < 6]
loud_short_planets
```

通常は1行で書きますが、3行に分けると構造がより明確になるかもしれません：

```python
[
    planet.upper() + '!' 
    for planet in planets 
    if len(planet) < 6
]
```

（SQLの類推を続けると、この3行はそれぞれSELECT、FROM、WHEREに対応すると考えることができます。）

左側の式は、技術的にはループ変数を含む必要はありません（含まないのはかなり珍しいですが）。以下の式はどのような結果になると思いますか？「output」ボタンを押して確認してみてください。

```python
[32 for planet in planets]
```

リスト内包表記を `min`、`max`、`sum` のような関数と組み合わせると、通常は数行のコードが必要な問題に対して、印象的な1行の解決策を生み出すことができます。

例えば、同じことをする以下の2つのコードセルを比較してみてください。

```python
def count_negatives(nums):
    """与えられたリスト内の負の数の個数を返す。
    
    >>> count_negatives([5, -1, -2, 0, 3])
    2
    """
    n_negative = 0
    for num in nums:
        if num < 0:
            n_negative = n_negative + 1
    return n_negative
```

リスト内包表記を使った解決策はこちらです：

```python
def count_negatives(nums):
    return len([num for num in nums if num < 0])
```

ずっと良いですよね？

コードの長さを最小限にすることだけが目標なら、3番目の解決策はさらに優れています！

```python
def count_negatives(nums):
    # 思い出してください：「ブール値と条件分岐」の演習で、Pythonでは
    # True + True + False + True のような計算が3になるという特徴を学びました。
    return sum([num < 0 for num in nums])
```

どの解決策が「最良」かは完全に主観的です。より少ないコードで問題を解決するのは常に良いことですが、[The Zen of Python](https://en.wikipedia.org/wiki/Zen_of_Python)の以下の行を念頭に置いておく価値があります：

> 読みやすさは重要である（Readability counts.）  
> 暗黙的より明示的な方が良い（Explicit is better than implicit.）

これらのツールを使ってコンパクトで読みやすいプログラムを作りましょう。ただし、選択を迫られたときは、他の人にとって理解しやすいコードを優先してください。
