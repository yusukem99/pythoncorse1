# はじめに

関数の入力値を変更すると、異なる出力が得られることをすでに見てきました。例えば、任意の数に5を足してその結果を返す `add_five()` 関数を考えてみましょう。`add_five(7)` は12（=7+5）を返し、`add_five(8)` は13（=8+5）を返します。入力が何であっても、関数が実行する処理は常に同じで、常に5を足します。

しかし、入力に応じて異なる処理を実行する関数が必要になることもあります。例えば、入力が10未満なら3を足し、10以上なら8を足す `add_three_or_eight()` 関数が必要になるかもしれません。この場合、`add_three_or_eight(1)` は4（=1+3）を返しますが、`add_three_or_eight(11)` は19（=11+8）を返します。このケースでは、関数が実行する処理は入力によって変わります。

このレッスンでは、条件と条件分岐を使って関数の動作を変更する方法を学びます。

# 条件

プログラミングにおいて、**条件**とは `True`（真）または `False`（偽）のいずれかになる文のことです。Pythonで条件を書く方法はたくさんありますが、最も一般的な方法の1つは、2つの異なる値を比較することです。例えば、2が3より大きいかどうかを確認できます。

```python
print(2 > 3)
```

Pythonはこれを `False` と判定します。2は3より大きくないからです。

条件を使って変数の値を比較することもできます。次のコードセルでは、`var_one` の値は1で、`var_two` の値は2です。条件式では、`var_one` が1より小さいかどうかを確認し（これは `False`）、`var_two` が `var_one` 以上かどうかを確認します（これは `True`）。

```python
var_one = 1
var_two = 2

print(var_one < 1)
print(var_two >= var_one)
```

**重要な注意**: 2つの値が等しいかどうかを確認するときは、`=` 記号ではなく、必ず `==` 記号を使用してください。
- `var_one==1` は `var_one` の値が1かどうかを確認しますが、
- `var_one=1` は `var_one` の値を1に設定します。

# 条件分岐

**条件分岐**は、条件を使って関数の動作を変更します。条件の値を確認し、条件が `True` と評価された場合、特定のコードブロックが実行されます。（条件が `False` の場合、そのコードは実行されません。）

以下のセクションでいくつかの例を見ていきましょう。

## "if" 文

最もシンプルな条件分岐は「if」文です。以下の `evaluate_temp()` 関数でその例を見ることができます。この関数は体温（摂氏）を入力として受け取ります。
- まず、`message` は `"Normal temperature"` に設定されます。
- 次に、`temp > 38` が `True`（つまり体温が38°Cより高い場合）であれば、メッセージは `"Fever!"` に更新されます。そうでなければ（`temp > 38` が `False` の場合）、メッセージは更新されません。
- 最後に、関数から `message` が返されます。

```python
def evaluate_temp(temp):
    # 初期メッセージを設定する
    message = "Normal temperature."
    # 体温が38より大きい場合のみメッセージの値を更新する
    if temp > 38:
        message = "Fever!"
    return message
```

次のコードセルでは、体温が37°Cの場合に関数を呼び出します。体温が38°C未満（`temp > 38` は `False` と評価される）なので、メッセージは `"Normal temperature"` のままです。

```python
print(evaluate_temp(37))
```

しかし、体温が39°Cの場合、38°Cより大きいため、メッセージは `"Fever!"` に更新されます。

```python
print(evaluate_temp(39))
```

インデントが2つのレベルあることに注目してください：
- 第1レベルのインデントは、関数内のコードブロックを常にインデントする必要があるためです。
- 第2レベルのインデントは、「if」文に属するコードブロックもインデントする必要があるためです。（後で見るように、「elif」文や「else」文のコードブロックもインデントする必要があります。）

return文は「if」文の下にインデントされていないため、`temp > 38` が `True` であっても `False` であっても常に実行されることに注意してください。

## "if ... else" 文

条件が `False` の場合にコードを実行するために「else」文を使用できます。「if」文の下のコードは条件が `True` の場合に実行され、「else」の下のコードは条件が `False` の場合に実行されます。

```python
def evaluate_temp_with_else(temp):
    if temp > 38:
        message = "Fever!"
    else:
        message = "Normal temperature."
    return message
```

この `evaluate_temp_with_else()` 関数は、`evaluate_temp()` 関数と同等の動作をします。

次のコードセルでは、体温が37°Cの場合にこの新しい関数を呼び出します。この場合、`temp > 38` は `False` と評価されるので、「else」文の下のコードが実行され、`Normal temperature.` というメッセージが返されます。

```python
print(evaluate_temp_with_else(37))
```

前の関数と同様に、「if」文と「else」文の後のコードブロックをインデントします。

## "if ... elif ... else" 文

「elif」（「else if」の略）を使って、複数の条件が真かどうかを確認できます。以下の関数は：
- まず `temp > 38` を確認します。これが真の場合、メッセージは `"Fever!"` に設定されます。
- メッセージがまだ設定されていない限り、次に `temp > 35` を確認します。これが真の場合、メッセージは `"Normal temperature."` に設定されます。
- それでもメッセージが設定されていない場合、「else」文により `"Low temperature."` メッセージが設定されます。

「elif」は「前の条件（例：`temp > 38`）は偽だったから、次にこの新しい条件（例：`temp > 35`）が真かどうか確認しよう！」と言っているようなものだと考えてください。

```python
def evaluate_temp_with_elif(temp):
    if temp > 38:
        message = "Fever!"
    elif temp > 35:
        message = "Normal temperature."
    else:
        message = "Low temperature."
    return message
```

以下のコードセルでは、`temp > 38` が `False` で `temp > 35` が `True` であるため、「elif」文の下のコードが実行されます。このコードが実行されると、関数は「else」文をスキップしてメッセージを返します。

```python
evaluate_temp_with_elif(36)
```

最後に、体温が35°C未満の場合を試してみましょう。「if」文と「elif」文の両方の条件が `False` と評価されるため、「else」文の中のコードブロックが実行されます。

```python
evaluate_temp_with_elif(34)
```

# 例 - 計算

これまでの例では、条件分岐を使って変数の値を設定する方法を決めていました。しかし、条件分岐を使って異なる計算を実行することもできます。

次の例では、税率が2段階しかない国に住んでいるとします。所得が12,000未満の人は25%の税金を払い、12,000以上の人は30%の税金を払います。以下の関数は、支払うべき税額を計算します。

```python
def get_taxes(earnings):
    if earnings < 12000:
        tax_owed = .25 * earnings
    else:
        tax_owed = .30 * earnings
    return tax_owed
```

次のコードセルでこの関数を使用します。

```python
ana_taxes = get_taxes(9000)
bob_taxes = get_taxes(15000)

print(ana_taxes)
print(bob_taxes)
```

それぞれのケースで、`get_taxes()` 関数を呼び出し、返された値を使って変数の値を設定しています。
- `ana_taxes` では、所得9,000の人が支払うべき税金を計算します。この場合、`earnings` を `9000` に設定して `get_taxes()` 関数を呼び出します。したがって、`earnings < 12000` は `True` であり、`tax_owed` は `.25 * 9000` に設定されます。そして `tax_owed` の値が返されます。
- `bob_taxes` では、所得15,000の人が支払うべき税金を計算します。この場合、`earnings` を `15000` に設定して `get_taxes()` 関数を呼び出します。したがって、`earnings < 12000` は `False` であり、`tax_owed` は `.30 * 15000` に設定されます。そして `tax_owed` の値が返されます。

別の例に進む前に、はじめに紹介した `add_three_or_eight()` 関数を覚えていますか？入力として数値を受け取り、入力が10未満なら3を足し、そうでなければ8を足します。この関数をどのように書くか分かりますか？答えが分かったら、下の「隠されたコードを表示」ボタンをクリックして解答を確認してください。

```python

def add_three_or_eight(number):
    if number < 10:
        result = number + 3
    else:
        result = number + 8
    return result
```

# 例 - 複数の "elif" 文

ここまで、関数内で「elif」を1回だけ使用する例を見てきました。しかし、「elif」文の数に制限はありません。例えば、次のコードブロックは、体重（キログラム）に基づいて子供に投与する薬の量（ミリリットル）を計算します。

注意：この関数は医療アドバイスとして使用すべきではなく、架空の薬を表しています。

```python
def get_dose(weight):
    # 体重が5.2kg未満の場合、投与量は1.25ml
    if weight < 5.2:
        dose = 1.25
    elif weight < 7.9:
        dose = 2.5
    elif weight < 10.4:
        dose = 3.75
    elif weight < 15.9:
        dose = 5
    elif weight < 21.2:
        dose = 7.5
    # 体重が21.2kg以上の場合、投与量は10ml
    else:
        dose = 10
    return dose
```

次のコードセルで関数を実行します。出力が理解できることを確認してください！
- この場合、「if」文は `False` であり、すべての「elif」文も `weight < 15.9` に到達するまで `False` と評価されます。`weight < 15.9` は `True` であるため、`dose` は5に設定されます。
- 「elif」文が `True` と評価されてコードブロックが実行されると、関数は残りのすべての「elif」文と「else」文をスキップします。スキップした後、残っているのはreturn文だけで、これが `dose` の値を返します。
- `elif` 文の順序はここでは重要です！文の順序を変更すると、非常に異なる結果になります。

```python
print(get_dose(12))
```
