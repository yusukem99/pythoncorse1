# ヘルプの使い方

前のチュートリアルで `abs` 関数を見ましたが、その機能を忘れてしまったらどうしますか？

`help()` 関数は、おそらくあなたが学ぶことができる最も重要なPython関数です。`help()` の使い方を覚えていれば、他のほとんどの関数を理解するための鍵を手にしていることになります。

以下に例を示します：

```python
help(round)
```

`help()` は2つの情報を表示します：

1. 関数のヘッダー `round(number, ndigits=None)`。この場合、`round()` は `number` と表現できる引数を取ることがわかります。さらに、`ndigits` と表現できる別の引数をオプションで指定できます。
2. 関数が何をするかの簡単な英語の説明。

**よくある落とし穴:** 関数を調べるときは、関数そのものの名前を渡すことを忘れないでください。関数を呼び出した結果を渡してはいけません。

`round()` 関数の*呼び出し結果*に対して help を実行するとどうなるでしょうか？以下のセルの出力を表示して確認してみましょう。

```python
help(round(-2.01))
```

Pythonはこのような式を内側から評価します。まず `round(-2.01)` の値を計算し、次にその式の出力に対するヘルプを提供します。

<small>（整数について言うことがたくさんあるようです！後でPythonのオブジェクト、メソッド、属性について学んだ後に、上のヘルプ出力がより理解できるようになります。）</small>

`round` は短いdocstring（ドキュメント文字列）を持つ非常にシンプルな関数です。`help` は `print` のような、より複雑で設定可能な関数を扱うときにさらに威力を発揮します。以下の出力が難解に見えても心配しないでください...今のところ、このヘルプから何か新しいことを見つけられるかどうか試してみてください。

```python
help(print)
```

注意深く見ていれば、print が `sep` という引数を取れること、そしてこれが他のすべての引数を出力するときにそれらの間に何を入れるかを指定するものであることがわかったかもしれません。

## 関数の定義

組み込み関数は便利ですが、それだけでは限界があり、いずれ自分自身の関数を定義する必要が出てきます。以下にシンプルな例を示します。

```python
def least_difference(a, b, c):
    diff1 = abs(a - b)
    diff2 = abs(b - c)
    diff3 = abs(a - c)
    return min(diff1, diff2, diff3)
```

これは `least_difference` という関数を作成し、3つの引数 `a`、`b`、`c` を取ります。

関数は `def` キーワードで始まるヘッダーから始まります。`:` の後のインデントされたコードブロックが、関数が呼び出されたときに実行されます。

`return` は関数に固有のもう1つのキーワードです。Pythonが `return` 文に出会うと、関数を直ちに終了し、右辺の値を呼び出し元に渡します。

ソースコードから `least_difference()` が何をするか理解できますか？わからない場合は、いくつかの例で試してみましょう：

```python
print(
    least_difference(1, 10, 100),
    least_difference(1, 10, 10),
    least_difference(5, 6, 7), # Pythonでは引数リストの末尾にカンマを付けることができます。便利ですね！
)
```

あるいは、`help()` 関数が何か教えてくれるかもしれません。

```python
help(least_difference)
```

Pythonは私のコードを読んで分かりやすい日本語の説明に変換するほど賢くはありません。しかし、関数を書くときに、**docstring（ドキュメント文字列）**と呼ばれるものに説明を書くことができます。

### Docstring（ドキュメント文字列）

```python
def least_difference(a, b, c):
    """Return the smallest difference between any two numbers
    among a, b and c.
    
    >>> least_difference(1, 5, -5)
    4
    """
    diff1 = abs(a - b)
    diff2 = abs(b - c)
    diff3 = abs(a - c)
    return min(diff1, diff2, diff3)
```

docstringは、関数のヘッダーの直後に来るトリプルクォートで囲まれた文字列です（複数行にわたることもあります）。`help()` を関数に対して呼び出すと、docstringが表示されます。

```python
help(least_difference)
```

> **補足：**
> docstringの最後の2行は、関数呼び出しとその結果の例です。（`>>>` はPythonの対話型シェルで使われるコマンドプロンプトへの参照です。）Pythonはこの例の呼び出しを実行するわけではなく、読者のために記載されているだけです。関数のdocstringに1つ以上の呼び出し例を含めるという慣習は広く普及しているとは言えませんが、関数を理解するのに非常に効果的です。実際の例として、[numpyの関数 `np.eye` のdocstring](https://github.com/numpy/numpy/blob/v1.14.2/numpy/lib/twodim_base.py#L140-L194)をご覧ください。

優れたプログラマーは、コードをすぐに捨てる予定でない限り（そのようなケースは稀ですが）、docstringを書きます。ですので、あなたもdocstringを書き始めましょう！

## 値を返さない関数

関数に `return` キーワードを含めなかった場合、どうなるでしょうか？

```python
def least_difference(a, b, c):
    """Return the smallest difference between any two numbers
    among a, b and c.
    """
    diff1 = abs(a - b)
    diff2 = abs(b - c)
    diff3 = abs(a - c)
    min(diff1, diff2, diff3)
    
print(
    least_difference(1, 10, 100),
    least_difference(1, 10, 10),
    least_difference(5, 6, 7),
)
```

Pythonではこのような関数を定義することが可能です。これらの関数を呼び出した結果は特殊な値 `None` になります。（これは他の言語における「null」の概念に似ています。）

`return` 文がなければ、`least_difference` は完全に無意味ですが、副作用を持つ関数は何も返さなくても有用な処理を行えます。すでに2つの例を見ています：`print()` と `help()` は何も返しません。画面にテキストを表示するという副作用のためだけに呼び出しています。その他の有用な副作用の例としては、ファイルへの書き込みや入力の変更があります。

```python
mystery = print()
print(mystery)
```

## デフォルト引数

`help(print)` を呼び出したとき、`print` 関数にはいくつかのオプション引数があることがわかりました。例えば、`sep` に値を指定して、出力する引数の間に特定の文字列を入れることができます：

```python
print(1, 2, 3, sep=' < ')
```

しかし、値を指定しない場合、`sep` はデフォルト値の `' '`（半角スペース1つ）として扱われます。

```python
print(1, 2, 3)
```

自分で定義する関数にデフォルト値を持つオプション引数を追加するのは、実はとても簡単です：

```python
def greet(who="Colin"):
    print("Hello,", who)
    
greet()
greet(who="Kaggle")
# （この場合、引数名を指定する必要はありません。曖昧さがないためです。）
greet("world")
```

## 関数に関数を適用する

最初は非常に抽象的に感じるかもしれませんが、これは強力な機能です。関数を他の関数の引数として渡すことができます。いくつかの例を見るとわかりやすくなるでしょう：

```python
def mult_by_five(x):
    return 5 * x

def call(fn, arg):
    """fnをargに対して呼び出す"""
    return fn(arg)

def squared_call(fn, arg):
    """fnをargに対して呼び出した結果に、さらにfnを呼び出す"""
    return fn(fn(arg))

print(
    call(mult_by_five, 1),
    squared_call(mult_by_five, 1), 
    sep='\n', # '\n' は改行文字 - 新しい行を開始します
)
```

他の関数を操作する関数は「高階関数」と呼ばれます。自分で書くのはまだ少し先になるかもしれません。しかし、Pythonには呼び出すと便利な高階関数が組み込まれています。

`max` 関数を使った興味深い例を見てみましょう。

デフォルトでは、`max` は引数の中で最大のものを返します。しかし、オプションの `key` 引数で関数を渡すと、`key(x)` を最大化する引数 `x` を返します（いわゆる「argmax」）。

```python
def mod_5(x):
    """xを5で割った余りを返す"""
    return x % 5

print(
    'Which number is biggest?',
    max(100, 51, 14),
    'Which number is the biggest modulo 5?',
    max(100, 51, 14, key=mod_5),
    sep='\n',
)
```
