## ステップ2: ランダムな値を作る — `random`

`random` モジュールを使うと、**毎回違う値** を作れます。これがあるとゲームに「予測不能」の楽しさが生まれます。

下のセルを **何回か実行** してみてください。実行するたびに違う数字が出るはずです。

```python
import random

secret = random.randint(1, 100)
print(f"今回のランダムな数字: {secret}")
```

**ポイント**:
- `random.randint(a, b)` は `a` 以上 `b` 以下のランダムな整数を返します
- セキュリティの世界では、ランダム性は **暗号・トークン生成・ペネトレーションテスト** などで重要な要素です

## ステップ3: メインチャレンジ — 暗証番号クラッカー

あなたはセキュリティ研究者です。ターゲットシステムの **3桁の暗証番号（1〜999）** を突破することがミッション。

サーバーは入力されたPINに対して、次のフィードバックを返します:

- 正解 → **アクセス成功**
- 小さすぎる → 「もっと大きい数」
- 大きすぎる → 「もっと小さい数」

**8回以内** に突破できるか？

下のセルを実行して、ゲームをプレイしてみてください。

```python
import random

target = random.randint(1, 999)
max_attempts = 8

print("=" * 40)
print("ターゲット: 機密サーバー")
print(f"PIN: 1〜999 のいずれか / 試行回数: {max_attempts}回")
print("=" * 40)

success = False
for attempt in range(1, max_attempts + 1):
    guess = int(input(f"[{attempt}/{max_attempts}] PINを入力: "))
    if guess == target:
        print(f"\n>>> ACCESS GRANTED <<<  突破成功！  ({attempt}回で正解)")
        success = True
        break
    elif guess < target:
        print("    -> もっと大きい数字")
    else:
        print("    -> もっと小さい数字")

if not success:
    print(f"\n>>> ACCESS DENIED <<<  侵入失敗。正解は {target} でした。")
```

**コードの解説**:
- `random.randint(1, 999)` で毎回違うPINを生成
- `for` ループで最大8回試行
- `if / elif / else` で大小を判定し、ヒントを返す
- 突破できたら `break` でループを抜ける
- ループが正常に終了したら（8回使い切ったら）失敗メッセージ

## ステップ5（発展・オプション）: パスワード強度チェッカーをWebで公開しよう

ここからは **余裕がある人向け** の発展課題です。

自分で作ったツールを **Webサイトとして公開** してみましょう。`Gradio` というライブラリを使うと、Pythonの関数を **数行のコードでWebアプリ** にできます。

**作るもの**: パスワードの強度を判定するWebツール。長さや使われている文字種を見て、安全性を5段階で評価します。

まずGradioをインストールします。

```python
!pip install --quiet gradio
```

次に、強度判定の関数を定義し、Gradioでラップして公開します。

下のセルを実行すると、**`https://...gradio.live` という公開URL** が表示されます。そのURLにアクセスすれば、世界中のどこからでもあなたのツールを使えます（72時間有効）。

クラスメイトにURLを送って、試してもらいましょう。

```python
import gradio as gr
import hashlib

def check_password(password):
    if not password:
        return "パスワードを入力してください"

    length = len(password)
    has_digit = any(c.isdigit() for c in password)
    has_upper = any(c.isupper() for c in password)
    has_lower = any(c.islower() for c in password)
    has_symbol = any(not c.isalnum() for c in password)

    score = sum([length >= 8, has_digit, has_upper, has_lower, has_symbol])
    ratings = ["非常に弱い", "弱い", "普通", "強い", "とても強い", "完璧"]
    rating = ratings[score]

    sha256 = hashlib.sha256(password.encode()).hexdigest()

    result = f"""
【強度評価】 {rating}  ({score} / 5 点)

長さ:           {length} 文字
数字を含む:     {"はい" if has_digit else "いいえ"}
大文字を含む:   {"はい" if has_upper else "いいえ"}
小文字を含む:   {"はい" if has_lower else "いいえ"}
記号を含む:     {"はい" if has_symbol else "いいえ"}

SHA-256 ハッシュ値:
{sha256}
"""
    return result

demo = gr.Interface(
    fn=check_password,
    inputs=gr.Textbox(label="パスワードを入力", type="password"),
    outputs=gr.Textbox(label="判定結果"),
    title="パスワード強度チェッカー",
    description="あなたのパスワードがどれくらい安全か診断します。（入力された値はサーバーに保存されません）",
)

demo.launch(share=True)
```

**発展のヒント**: `check_password` 関数を改造して、独自の判定ロジックを足してみてください。例:

- 「password」「123456」など、よくある弱いパスワードを検出してエラーを返す
- 同じ文字が連続して何回出てくるかチェック
- 推定クラック時間を概算して表示

関数を書き換えたら、もう一度セルを実行すると、URLは変わらずに中身だけ更新されます。
