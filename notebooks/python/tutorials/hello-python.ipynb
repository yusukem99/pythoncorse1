# Hello, Python!

Pythonはイギリスのコメディグループ[モンティ・パイソン](https://en.wikipedia.org/wiki/Monty_Python)にちなんで名付けられました。そこで、最初のPythonプログラムは彼らの[スパム](https://bit.ly/3MwwjSs)のコントへのオマージュにしましょう。

楽しみとして、以下のコードを読んで、実行するとどうなるか予想してみてください。（わからなくても大丈夫です！）

その後、「output」ボタンをクリックしてプログラムの結果を確認しましょう。

````python
spam_amount = 0
print(spam_amount)

# スパム、卵、スパム、スパム、ベーコン、スパムを注文（スパムをさらに4人前追加）
spam_amount = spam_amount + 4

if spam_amount > 0:
    print("But I don't want ANY spam!")

viking_song = "Spam " * spam_amount
print(viking_song)
````

ここには解説すべきことがたくさんあります！このちょっとおかしなプログラムは、Pythonコードがどのように見え、どのように動作するかの重要な側面を多く示しています。コードを上から順に見ていきましょう。

````python
spam_amount = 0
````

**変数の代入:** ここでは `spam_amount` という変数を作成し、**代入演算子**と呼ばれる `=` を使って値 0 を代入しています。

> **注意**: 他の言語（JavaやC++など）でプログラミングした経験がある方は、Pythonでは以下のことが*不要*であることに気づくかもしれません：
- `spam_amount` に代入する前に「宣言」する必要がありません
- `spam_amount` がどのような型の値を参照するかをPythonに伝える必要がありません。実際、`spam_amount` を文字列やブール値など、まったく別の種類のものに再代入することも可能です。

````python
print(spam_amount)
````

**関数呼び出し:** `print` は、渡された値を画面に表示するPythonの関数です。関数を呼び出すには、関数名の後に括弧を付け、その括弧の中に関数への入力（**引数**）を入れます。

````python
# スパム、卵、スパム、スパム、ベーコン、スパムを注文（スパムをさらに4人前追加）
spam_amount = spam_amount + 4
````

上の最初の行は**コメント**です。Pythonでは、コメントは `#` 記号で始まります。

次に、再代入の例を見てみましょう。既存の変数の値を再代入する方法は、変数を作成するときとまったく同じで、`=` 代入演算子を使います。

この場合、`spam_amount` に代入している値は、以前の値に対する簡単な算術演算を含んでいます。Pythonはこの行に出会うと、`=` の右辺の式を評価し（0 + 4 = 4）、その値を左辺の変数に代入します。

````python
if spam_amount > 0:
    print("But I don't want ANY spam!")

viking_song = "Spam Spam Spam"
print(viking_song)
````

「条件分岐」については後のレッスンで詳しく説明しますが、プログラミングの経験がなくても、このコードが何をするかはおそらく想像がつくでしょう。Pythonは読みやすさとシンプルさで高く評価されています。

`if` に属するコードをどのように示しているかに注目してください。`"But I don't want ANY spam!"` は `spam_amount` が正の場合にのみ出力されるべきです。しかし、後のコード（`print(viking_song)` など）は何があっても実行されるべきです。私たち（そしてPython）はどうやってそれを判別しているのでしょうか？

`if` 行の最後にあるコロン（`:`）は、新しい**コードブロック**が始まることを示しています。その後のインデントされた行がそのコードブロックの一部です。
> **注意**: 他の言語でプログラミングした経験がある方は、コードブロックの始まりと終わりを `{` 波括弧 `}` で示す言語があることをご存知かもしれません。Pythonが意味のある空白を使うことは、他の言語に慣れたプログラマーにとって驚きかもしれませんが、実際にはコードブロックのインデントを強制しない言語よりも、一貫性があり読みやすいコードにつながります。

`viking_song` を扱う後の行は追加の4つのスペースでインデントされていないため、`if` のコードブロックの一部ではありません。インデントされたコードブロックの例は、関数の定義やループの使用時にさらに見ていきます。

このコードには、Pythonの**文字列**も初めて登場しています：

```python
"But I don't want ANY spam!"
```

文字列はダブルクォートまたはシングルクォートで囲むことができます。（ただし、この文字列にはシングルクォート文字が*含まれている*ため、注意しないとシングルクォートで囲もうとするとPythonが混乱する可能性があります。）

````python
viking_song = "Spam " * spam_amount
print(viking_song)
````

`*` 演算子は2つの数値を掛け合わせるために使えますが（`3 * 3` は 9 になります）、文字列に数値を掛けて、その回数だけ繰り返したバージョンを得ることもできます。Pythonには、`*` や `+` のような演算子が、適用される対象の種類によって異なる意味を持つという、ちょっとした便利なトリックがいくつかあります。（これを技術的には[演算子のオーバーロード](https://en.wikipedia.org/wiki/Operator_overloading)と呼びます。）

## Pythonにおける数値と算術演算

上で数値を含む変数の例をすでに見ました：

````python
spam_amount = 0
````

「数値」はこの種のものを指す日常的な名前としては十分ですが、もう少し技術的に言えば、`spam_amount` がどのような型なのかPythonに聞いてみることができます：

````python
type(spam_amount)
````

これは `int`（整数、integerの略）です。Pythonでよく使うもう1つの数値の種類があります：

````python
type(19.95)
````

`float` は小数点を持つ数値で、重さや割合などを表すのに非常に便利です。

`type()` は（`print()` に続いて）2番目に登場した組み込み関数ですが、これも覚えておくと良い関数です。Pythonに「これは何の種類のものですか？」と聞けるのはとても便利です。

数値に対して自然にやりたくなることは算術演算です。加算の `+` 演算子と乗算の `*` 演算子はすでに見ました。Pythonは電卓の残りの基本ボタンもすべてカバーしています：

| 演算子        | 名前           | 説明                                                   |
|--------------|----------------|--------------------------------------------------------|
| ``a + b``    | 加算           | ``a`` と ``b`` の和                                     |
| ``a - b``    | 減算           | ``a`` と ``b`` の差                                     |
| ``a * b``    | 乗算           | ``a`` と ``b`` の積                                     |
| ``a / b``    | 真の除算       | ``a`` を ``b`` で割った商                                |
| ``a // b``   | 切り捨て除算   | ``a`` を ``b`` で割った商の小数部分を切り捨てたもの        |
| ``a % b``    | 剰余           | ``a`` を ``b`` で割った余り                              |
| ``a ** b``   | 累乗           | ``a`` の ``b`` 乗                                       |
| ``-a``       | 符号反転       | ``a`` の負の値                                          |

<span style="display:none"></span>

ここで興味深いのは、電卓には通常除算のボタンが1つしかないのに対し、Pythonでは2種類の除算ができることです。「真の除算」は基本的に電卓と同じ動作をします：

````python
print(5 / 2)
print(6 / 2)
````

結果は常に `float` になります。

`//` 演算子は、結果を次の整数に切り捨てた値を返します。

````python
print(5 // 2)
print(6 // 2)
````

これがどこで役に立つか思いつきますか？コーディング課題ですぐに例を見ることになります。

### 演算の順序

小学校で習った算術には、演算が評価される順序に関する規則があります。**PEMDAS**（**P**arentheses（括弧）、**E**xponents（累乗）、**M**ultiplication（乗算）/**D**ivision（除算）、**A**ddition（加算）/**S**ubtraction（減算））という語呂合わせで覚えている方もいるでしょう。

Pythonも同様のルールに従って、どの計算を先に行うかを決定します。ほとんどの場合、直感的に理解できるものです。

````python
8 - 3 + 2
````

````python
-3 + 4 * 2
````

デフォルトの演算順序が望む結果にならないこともあります：

````python
hat_height_cm = 25
my_height_cm = 190
# 帽子をかぶったときの身長はメートルで何メートル？
total_height_meters = hat_height_cm + my_height_cm / 100
print("Height in meters =", total_height_meters, "?")
````

ここでは括弧が便利です。括弧を追加して、Pythonに部分式を任意の順序で評価させることができます。

````python
total_height_meters = (hat_height_cm + my_height_cm) / 100
print("Height in meters =", total_height_meters)
````

### 数値を扱う組み込み関数

`min` と `max` はそれぞれ引数の最小値と最大値を返します...

````python
print(min(1, 2, 3))
print(max(1, 2, 3))
````

`abs` は引数の絶対値を返します：

````python
print(abs(32))
print(abs(-32))
````

Pythonの2つの主要な数値型の名前であるだけでなく、`int` と `float` は引数を対応する型に変換する関数としても呼び出すことができます：

````python
print(float(10))
print(int(3.33))
# 文字列に対しても呼び出せます！
print(int('807') + 1)
````
