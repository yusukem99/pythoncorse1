- プレイヤーに表向きのカードが2枚配られます。ディーラーには表向きのカードが1枚配られます。
- プレイヤーは好きなだけ追加のカードを要求（「ヒット」）できます。カードの合計が21を超えると、その時点で負けとなります。
- その後、ディーラーは以下のいずれかになるまで自分にカードを配ります：

    - ディーラーのカードの合計が17以上になった場合。プレイヤーの合計がディーラーより大きければプレイヤーの勝ち。それ以外はディーラーの勝ち（同点の場合もディーラーの勝ち）。
    
カードの合計を計算する際、Jack、Queen、Kingは10として数えます。エースは1または11として数えることができます（プレイヤーの「合計」とは、21を超えない範囲で最大の合計のことです。例：A+8 = 19、A+8+8 = 17）
