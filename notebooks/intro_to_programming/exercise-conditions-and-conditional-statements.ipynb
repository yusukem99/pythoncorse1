# 🌶️ 問題 5

メキシコでは、飽和脂肪、トランス脂肪、砂糖、ナトリウム、および/またはカロリーが高い食品や飲料に、消費者が健康的な食品選択をするための警告ラベルが表示されます。

例えば、下の画像にある[クッキーの箱](https://world.openfoodfacts.org/product/7501000673209/florentinas-gamesa)には、2つのラベルが表示されています（右上隅）：
- EXCESO CALORÍAS（日本語で「カロリー過多」）
- EXCESO AZÚCARES（日本語で「糖分過多」）

<center><img src="https://storage.googleapis.com/kaggle-media/learn/images/VXYKHnM.png" alt="drawing" width="500"/></center>

この問題では、食品の栄養情報を受け取り、必要な警告ラベルを表示する関数 `get_labels()` を扱います。この関数はいくつかの入力を受け取ります：
- `food_type` = `"solid"`（固体）または `"liquid"`（液体）のいずれか
- `serving_size` = 1食分のサイズ（固体の場合はグラム、液体の場合はミリリットル）
- `calories_per_serving` = 1食分のカロリー
- `saturated_fat_g` = 1食分の飽和脂肪のグラム数
- `trans_fat_g` = 1食分のトランス脂肪のグラム数
- `sodium_mg` = 1食分のナトリウムのmg数
- `sugars_g` = 1食分の砂糖のグラム数

ここでは、`excess_sugar()` や `excess_saturated_fat()` などの関数の詳細を共有していないため、一部のコードは馴染みがないかもしれません。しかし概要としては、これらは食品に砂糖や飽和脂肪がそれぞれ過剰に含まれていると判断された場合に `True` の値を返す関数です。これらの関数は `get_labels()` 関数内で使用され、過剰（砂糖や飽和脂肪だけでなく、トランス脂肪、ナトリウム、カロリーも含む）がある場合に対応するラベルを表示します。

```python
# get_labels を動作させるために必要な関数をインポートする
from learntools.intro_to_programming.ex4q5 import excess_sugar, excess_saturated_fat, excess_trans_fat, excess_sodium, excess_calories

def get_labels(food_type, serving_size, calories_per_serving, saturated_fat_g, trans_fat_g, sodium_mg, sugars_g):
    # 判定結果に基づいてメッセージを表示する
    if excess_sugar(sugars_g, calories_per_serving) == True:
        print("EXCESO AZÚCARES / EXCESS SUGAR")
    if excess_saturated_fat(saturated_fat_g, calories_per_serving) == True:
        print("EXCESO GRASAS SATURADAS / EXCESS SATURATED FAT")
    if excess_trans_fat(trans_fat_g, calories_per_serving) == True:
        print("EXCESO GRASAS TRANS / EXCESS TRANS FAT")
    if excess_sodium(calories_per_serving, sodium_mg) == True:
        print("EXCESO SODIO / EXCESS SODIUM")
    if excess_calories(food_type, calories_per_serving, serving_size) == True:
        print("EXCESO CALORÍAS / EXCESS CALORIES")
```

次のコードセルは、`get_labels()` を使って食品に必要な警告ラベルを取得する方法を示しています。まず[ボローニャソーセージ](https://world.openfoodfacts.org/product/4099100179378/bologna)から始めます。[こちらの画像](https://storage.googleapis.com/kaggle-media/learn/images/Cfcx72e.png)にすべての栄養情報があります。この食品については、

- `sodium_mg = 400`（1食分のナトリウムは400mg）
- `sugars_g = 1`（栄養成分表示では<1gとなっていますが、安全のため1食分1グラムに切り上げます）

これらの値をすべて関数に渡すことで、警告ラベルを表示できます。
