# リスト（list）

Pythonのリストは、値の順序付きの並び（シーケンス）を表します。以下はリストの作成例です：

```python
primes = [2, 3, 5, 7]
```

リストには他の型のデータも入れることができます：

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
```

リストの中にリストを入れることもできます（入れ子のリスト）：

```python
hands = [
    ['J', 'Q', 'K'],
    ['2', '2', '2'],
    ['6', 'A', 'K'], # （最後の要素の後のカンマは省略可能）
]
# （1行で書くこともできますが、読みにくくなることがあります）
hands = [['J', 'Q', 'K'], ['2', '2', '2'], ['6', 'A', 'K']]
```

リストには異なる型のデータを混在させることもできます：

```python
my_favourite_things = [32, 'raindrops on roses', help]
# （そう、Pythonのhelp関数は間違いなく私のお気に入りの1つです）
```

## インデックス（Indexing）

角括弧を使ってリストの個々の要素にアクセスできます。

太陽に最も近い惑星はどれでしょうか？Pythonは*ゼロベースの*インデックスを使用するので、最初の要素のインデックスは0です。

```python
planets[0]
```

次に近い惑星は？

```python
planets[1]
```

太陽から*最も遠い*惑星はどれでしょうか？

リストの末尾の要素には、-1から始まる負のインデックスでアクセスできます：

```python
planets[-1]
```

```python
planets[-2]
```

## スライス（Slicing）

最初の3つの惑星は何でしょうか？*スライス*を使って答えることができます：

```python
planets[0:3]
```

`planets[0:3]` は、`planets` のインデックス0から始まり、インデックス3を*含まない*までの要素を要求する方法です。

開始インデックスと終了インデックスはどちらもオプションです。開始インデックスを省略すると0とみなされます。したがって、上の式は次のように書き換えることもできます：

```python
planets[:3]
```

終了インデックスを省略すると、リストの長さとみなされます。

```python
planets[3:]
```

つまり、上の式は「インデックス3以降のすべての惑星をください」という意味です。

スライスでも負のインデックスを使うことができます：

```python
# 最初と最後を除くすべての惑星
planets[1:-1]
```

```python
# 最後の3つの惑星
planets[-3:]
```

## リストの変更

リストは「ミュータブル（変更可能）」です。つまり、「その場で」変更できます。

リストを変更する方法の1つは、インデックスまたはスライス式に代入することです。

例えば、火星（Mars）の名前を変更したいとしましょう：

```python
planets[3] = 'Malacandra'
planets
```

うーん、それはちょっと言いにくいですね。最初の3つの惑星の名前を短くして埋め合わせましょう。

```python
planets[:3] = ['Mur', 'Vee', 'Ur']
print(planets)
# これはバカバカしかったですね。元の名前に戻しましょう
planets[:4] = ['Mercury', 'Venus', 'Earth', 'Mars',]
```

## リスト関数

Pythonにはリストを扱うための便利な関数がいくつかあります。

`len` はリストの長さを返します：

```python
# 惑星はいくつある？
len(planets)
```

`sorted` はリストのソート済みバージョンを返します：

```python
# 惑星をアルファベット順にソート
sorted(planets)
```

`sum` は期待通りの動作をします：

```python
primes = [2, 3, 5, 7]
sum(primes)
```

以前、`min` と `max` を使って複数の引数の最小値や最大値を取得しました。しかし、リストを1つの引数として渡すこともできます。

```python
max(primes)
```

## 余談：オブジェクト

ここまで「オブジェクト」という用語をたくさん使ってきました。Pythonでは*すべて*がオブジェクトだと聞いたことがあるかもしれません。それはどういう意味でしょうか？

簡単に言うと、オブジェクトは何かを持ち運んでいます。そしてその中身にはPythonのドット構文を使ってアクセスします。

例えば、Pythonの数値は `imag` という関連する変数を持っていて、虚数部を表します。（非常に特殊な数学をしない限り、これを使う必要はおそらくないでしょう。）

```python
x = 12
# xは実数なので、虚数部は0
print(x.imag)
# 複素数の作り方です。興味があれば参考に：
c = 12 + 3j
print(c.imag)
```

オブジェクトが持ち運ぶものには関数も含まれます。オブジェクトに付属する関数は**メソッド**と呼ばれます。（`imag` のような、関数でないものは*属性（アトリビュート）*と呼ばれます。）

例えば、数値には `bit_length` というメソッドがあります。これもドット構文でアクセスします：

```python
x.bit_length
```

実際に呼び出すには、括弧を付けます：

```python
x.bit_length()
```

> **補足：** 演習をやっていれば、実はすでにメソッドを呼び出しています。演習ノートブックの `q1`、`q2`、`q3` などはすべて、`check`、`hint`、`solution` というメソッドを持つオブジェクトです。

`help` 関数に関数を渡せるように（例：`help(max)`）、メソッドも渡すことができます：

```python
help(x.bit_length)
```

上の例はかなりマニアックなものでした。これまで見てきたオブジェクトの型（数値、関数、ブール値）には、実際に使うことのある属性やメソッドはほとんどありません。

しかし、リストには頻繁に使うメソッドがいくつもあることがわかります。

## リストのメソッド

`list.append` はリストの末尾に要素を追加してリストを変更します：

```python
# 冥王星だって惑星だ！
planets.append('Pluto')
```

上のセルに出力がないのはなぜでしょうか？`help(planets.append)` を呼び出してドキュメントを確認してみましょう。

> **補足：** `append` は `planets` だけでなく、list型の*すべての*オブジェクトが持つメソッドです。そのため `help(list.append)` と呼ぶこともできます。しかし、`help(append)` と呼ぼうとすると、「append」という変数が存在しないとPythonに怒られます。「append」という名前はリストの中にのみ存在し、`max` や `len` のような組み込み関数のような独立した名前としては存在しません。

```python
help(planets.append)
```

`-> None` の部分は、`list.append` が何も返さないことを示しています。しかし、`planets` の値を確認すると、メソッド呼び出しが `planets` の値を変更したことがわかります：

```python
planets
```

`list.pop` はリストの最後の要素を削除して返します：

```python
planets.pop()
```

```python
planets
```

### リストの検索

地球は惑星の順番で何番目でしょうか？`list.index` メソッドを使ってインデックスを取得できます。

```python
planets.index('Earth')
```

3番目です（つまりインデックス2 - 0始まりのインデックスであることをお忘れなく！）。

冥王星（Pluto）は何番目のインデックスにありますか？

```python
planets.index('Pluto')
```

あ、そうでした...

このような不愉快な驚きを避けるために、`in` 演算子を使ってリストに特定の値が含まれているかどうかを判定できます：

```python
# 地球は惑星ですか？
"Earth" in planets
```

```python
# Calbefraquesは惑星ですか？
"Calbefraques" in planets
```

ここでは紹介しきれなかった興味深いリストメソッドがまだいくつかあります。特定のオブジェクトに付属するすべてのメソッドと属性について知りたい場合は、オブジェクト自体に対して `help()` を呼び出すことができます。例えば、`help(planets)` は*すべての*リストメソッドについて教えてくれます：

```python
help(planets)
```

「output」ボタンをクリックして完全なヘルプページを確認してみてください。リストには `__eq__` や `__iadd__` のような見慣れない名前のメソッドがたくさんあります。今のところあまり気にしなくて大丈夫です。（このようなメソッドを直接呼び出すことはおそらくありません。しかし、インデックスや比較演算子などの構文を使ったときに裏で呼び出されています。）最も興味深いメソッドはリストの下の方にあります（`append`、`clear`、`copy` など）。

## タプル（Tuples）

タプルはリストとほぼ同じです。違いは2つだけです。

**1:** 作成時の構文が角括弧ではなく丸括弧を使います。

```python
t = (1, 2, 3)
```

```python
t = 1, 2, 3 # 上と同等
t
```

**2:** 変更できません（*イミュータブル（不変）*です）。

```python
t[0] = 100
```

タプルは複数の戻り値を持つ関数でよく使われます。

例えば、floatオブジェクトの ``as_integer_ratio()`` メソッドは、分子と分母をタプルの形で返します：

```python
x = 0.125
x.as_integer_ratio()
```

これらの複数の戻り値は、次のように個別に変数に代入できます：

```python
numerator, denominator = x.as_integer_ratio()
print(numerator / denominator)
```

最後に、2つの変数を入れ替えるという、Pythonの古典的な裏ワザの仕組みがわかります！

```python
a = 1
b = 0
a, b = b, a
print(a, b)
```
