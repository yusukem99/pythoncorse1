# インポート（import）

これまで、Python言語に組み込まれた型や関数について学んできました。

しかし、Python（特にデータサイエンティストにとって）の最も素晴らしい点の一つは、高品質なカスタムライブラリが膨大に存在することです。

これらのライブラリの一部は「標準ライブラリ」に含まれており、Pythonが動作する環境であればどこでも利用できます。その他のライブラリも、Pythonに常に同梱されているわけではありませんが、簡単にインストールできます。

いずれの場合も、**import** を使ってそのコードにアクセスします。

まずは標準ライブラリから `math` をインポートする例から始めましょう。

````python
import math

print("It's math! It has type {}".format(type(math)))
````

`math` はモジュールです。モジュールとは、誰かが定義した変数の集まり（*名前空間*とも言います）です。組み込み関数 `dir()` を使って、`math` の中にあるすべての名前を確認できます。

````python
print(dir(math))
````

ドット構文を使ってこれらの変数にアクセスできます。中には `math.pi` のような単純な値を参照するものもあります：

````python
print("pi to 4 significant digits = {:.4}".format(math.pi))
````

しかし、モジュール内で見つかるもののほとんどは `math.log` のような関数です：

````python
math.log(32, 2)
````

もちろん、`math.log` が何をするかわからない場合は、`help()` を呼び出すことができます：

````python
help(math.log)
````

モジュール自体に対して `help()` を呼び出すこともできます。これにより、モジュール内の*すべての*関数と値のドキュメントが結合して表示されます（モジュールの概要説明も含みます）。「output」ボタンをクリックして `math` の完全なヘルプページを確認してみてください。

````python
help(math)
````

### その他のインポート構文

`math` の関数を頻繁に使うことがわかっている場合、入力を省くために短いエイリアスでインポートできます（この場合「math」はすでにかなり短いですが）。

````python
import math as mt
mt.pi
````

> Pandas、Numpy、Tensorflow、Matplotlibなどの人気ライブラリでこのようなコードを見たことがあるかもしれません。例えば、`import numpy as np` や `import pandas as pd` とするのが一般的な慣習です。

`as` は単にインポートしたモジュールの名前を変更するだけです。以下のようなことをするのと同等です：

````python
import math
mt = math
````

`math` モジュール内のすべての変数を直接参照できたら便利だと思いませんか？つまり、`math.pi` や `mt.pi` の代わりに、単に `pi` と書けたら？良いニュースです：それが可能です。

````python
from math import *
print(pi, log(32, 2))
````

`import *` はモジュールのすべての変数に直接アクセスできるようにします（ドット付きのプレフィックスなしで）。

悪いニュース：純粋主義者の中にはこれに文句を言う人もいるかもしれません。

さらに悪いニュース：彼らにはある程度正当な理由があります。

````python
from math import *
from numpy import *
print(pi, log(32, 2))
````

何が起こったのでしょうか？さっきまでは動いていたのに！

このような「スターインポート」は、時として奇妙でデバッグが困難な状況を引き起こすことがあります。

この場合の問題は、`math` モジュールと `numpy` モジュールの両方に `log` という関数があるのに、それぞれ異なる動作をすることです。`numpy` を後にインポートしているため、`numpy` の `log` が `math` からインポートした `log` 変数を上書き（「シャドウイング」）してしまいます。

良い妥協案は、各モジュールから必要なものだけを具体的にインポートすることです：

````python
from math import log, pi
from numpy import asarray
````

### 演算子のオーバーロード

以下の式の値はいくつでしょうか？

````python
[3, 4, 1, 2, 2, 1] + 10
````

馬鹿な質問ですね。もちろんエラーになります。

しかし、これはどうでしょう...

````python
rolls + 10
````

Pythonは `+`、`<`、`in`、`==`、またはインデックスやスライスの角括弧など、コア構文の動作を厳密に管理していると思うかもしれません。しかし実際には、非常に自由なアプローチを取っています。新しい型を定義するとき、加算の動作を決めたり、その型のオブジェクトが他の何かと等しいとはどういう意味かを定義したりできます。

リストの設計者は、リストに数値を足すことを許可しないことにしました。`numpy` 配列の設計者は別の道を選びました（配列の各要素に数値を加算する）。

`numpy` 配列がPythonの演算子と予想外の（少なくともリストとは異なる）やり取りをする例をもう少し見てみましょう。

````python
# サイコロの目が3以下のインデックスはどれ？
rolls <= 3
````

````python
xlist = [[1,2,3],[2,4,6],]
# 2次元配列を作成
x = numpy.asarray(xlist)
print("xlist = {}\nx =\n{}".format(xlist, x))
````

````python
# numpy配列の2行目の最後の要素を取得
x[1,-1]
````

````python
# ネストしたリストの2番目のサブリストの最後の要素を取得？
xlist[1,-1]
````

numpyの `ndarray` 型は多次元データの操作に特化しているため、独自のインデックスロジックを定義しており、各次元のインデックスをタプルで指定できます。

### 1 + 1 が 2 にならないとき

もっと奇妙なことも起こりえます。ディープラーニングでよく使われるPythonライブラリのtensorflowを聞いたことがある（あるいは使ったことがある）かもしれません。tensorflowは演算子のオーバーロードを広範に活用しています。

````python
import tensorflow as tf
# 値が1の定数を2つ作成
a = tf.constant(1)
b = tf.constant(1)
# 足し合わせると...
a + b
````

`a + b` は2ではありません。tensorflowのドキュメントを引用すると...

> `Operation` の出力の1つへのシンボリックハンドルです。その演算の出力値を保持するのではなく、TensorFlowの `tf.Session` でそれらの値を計算する手段を提供します。



このようなことが可能であること、そしてライブラリが演算子のオーバーロードを非自明な、あるいは魔法のような方法で使うことがあるということを認識しておくことが重要です。

int、文字列、リストに適用されたPythonの演算子の動作を理解していても、tensorflowの `Tensor`、numpyの `ndarray`、pandasの `DataFrame` に適用されたときに何をするか、すぐに理解できる保証はありません。

例えば、DataFrameに少し触れたことがあれば、以下のような式が直感的で魅力的に見え始めます：

```python
# 南米で人口100万以上の行を取得
df[(df['population'] > 10**6) & (df['continent'] == 'South America')]
```

しかし、なぜこれが動くのでしょうか？上の例にはおよそ**5つ**の異なるオーバーロードされた演算子が含まれています。それぞれの演算は何をしているのでしょうか？うまくいかなくなったときには、答えを知っていると助かります。

#### 仕組みが気になりますか？

オブジェクトに対して `help()` や `dir()` を呼び出して、ダブルアンダースコアの名前がたくさん出てきて、一体何なんだろうと思ったことはありませんか？

````python
print(dir(list))
````

これは実は演算子のオーバーロードと直接関係しています。

Pythonプログラマーが自分の型に対する演算子の動作を定義したいとき、`__lt__`、`__setattr__`、`__contains__` のように、2つのアンダースコアで始まり2つのアンダースコアで終わる特別な名前のメソッドを実装します。一般的に、このダブルアンダースコア形式の名前はPythonにとって特別な意味を持ちます。

例えば、`x in [1, 2, 3]` という式は、実はリストの `__contains__` メソッドを裏で呼び出しています。これは（見た目は悪いですが）`[1, 2, 3].__contains__(x)` と同等です。

さらに詳しく知りたい場合は、[Pythonの公式ドキュメント](https://docs.python.org/3.4/reference/datamodel.html#special-method-names)を確認してみてください。多くの特殊な「アンダースコア」メソッドが説明されています。
